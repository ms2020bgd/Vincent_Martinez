```python
df = pd.read_csv("https://raw.githubusercontent.com/fspot/INFMDI-721/master/lesson5/products.csv", sep=';')
df.shape
```

```python
# Cleaning the price
splitprice = df["price"].str.split(" ", n = 1, expand = True) 
df.price = splitprice[0]
df["currency"] = splitprice[1]
df.head(50)

```

```python
# Finding currency by IP

import CurrencyByCountry;

def getCurrencyFromIP(ip):

    try:
        details = ipHandler.getDetails(ip)
        return CurrencyByCountry.CURRENCIES_BY_COUNTRY_CODE[details.country][0]
    except:
        return None
    

#df.where(df.currency.isnull(), getCurrencyFromIP(df.ip_address))
df.currency = df.loc[df.currency.isnull(),"ip_address"].apply(getCurrencyFromIP)


```

```python
# in case of None currency, drop
df.dropna(subset=['currency'], inplace=True)
```

```python

exchangeRate = requests.get("http://data.fixer.io/api/latest?access_key=af365b53c170f61e086b8570ee2481bf&format=1").json()["rates"]
exchangeRate['EUR'] = 1
df.price = pd.to_numeric(df.price)

df.price = df.apply(lambda x: x['price'] / exchangeRate[x['currency']], axis=1)
```

```python
# remove all punctuation from the info
splits = df.infos.str.lower().str.replace(',','').str.replace(':','').str.replace('and','').str.replace('may','').str.replace('contains','').str.replace('contain','').str.replace("ingredients",'').str.split(" ")
```

```python
dummies = pd.get_dummies(splits.explode()) # need 0.25 pandas
dummies = dummies.groupby(dummies.index).sum()[:]
dummies = dummies.rename(columns={"": "remove"}).drop(columns="remove")[:].astype('bool')
```

```python
# Cleaning df and concate the dataframe

df=df.drop(columns=["ip_address", "currency", "infos"])
```

```python
df = df.join(dummies)
```
