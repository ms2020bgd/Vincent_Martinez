```python
import pandas as pd
```

```python
population = pd.read_csv("population_communes.csv")
population.head()

surface = pd.read_csv("surface_departements.csv")
surface.head()
```

```python
pop_sum = population.groupby(["variable", "Code département", "Code région"])["value"].sum()
pop_sum = pop_sum.reset_index()
pop_sum = pop_sum.pivot_table(values="value", index=["Code département", "Code région"], columns="variable") # Pivot_table >>> pivot. Comme ca, on peut faire du multi-index
pop_sum.head()
pop_sum = pop_sum.reset_index()
pop_sum['pct_pop_municipale'] = pop_sum["Population municipale"]/pop_sum["Population totale"]
#pop_sum["Population municipale"] = pop_sum[pop_sum["variable"]=="Population municipale"]["value"]
#pop_sum["Population totale"] = pop_sum[pop_sum["variable"]=="Population totale"]["value"]
```

```python
pop_sum = pop_sum.merge(surface, left_on="Code département", right_on="code_insee")
```

```python
pop_sum["Population Région"] = pop_sum.groupby("Code région").transform('sum').reset_index()["Population totale"] # Beaucoup plus simple le transform plutot que sum. On evite un join apres
```

```python
pop_sum['Densité'] = pop_sum["Population totale"]/pop_sum["surf_km2"]
```

```python
pop_sum['Part région'] = 100*pop_sum["Population totale"]/pop_sum["Population Région"]
```
